```python
work_years=[1,2,3,4]
ml_engineers_income=[10000,12000,14000,16000]
sde_income=[8000,11000,10000,17000]
income_df= pd.DataFrame({"Years at work": work_years, "Ml_Engineers_Income":ml_engineers_income,"SDE_income":sde_income})
```

```python
income_df.head()
```

```python
(ggplot()+
 plotnine.geom_line(data=income_df, mapping=aes(x="Years at work",y="Ml_Engineers_Income", color="'blue'"))+
  plotnine.geom_line(data=income_df, mapping=aes(x="Years at work",y="SDE_income", color="'red'"))+
 plotnine.labs(title="Years VS Income")+
 plotnine.scale_color_identity(guide='legend',name="Legend", breaks=['blue','red'],labels=['ML_Engineers Income','SDE_Income']))
```

```python
df= sns.load_dataset('titanic')
df.head()
```

```python
(plotnine.ggplot(data=df.dropna(),mapping= aes(x='pclass',y='fare',fill='sex'))+
 plotnine.geom_col()+
 plotnine.labs(title="Bar plot of passenger class"))
```

```python
scatter_plot= (ggplot(data= df.dropna(), mapping= aes(x='pclass', y='fare',color= 'sex'))+
                     plotnine.labs(title="Scatter plot of passenger class")+
               plotnine.geom_point(shape='o',size=7,color="green", alpha=0.7))
```

```python
scatter_plot
```

```python
hist= (ggplot(data=df.dropna(),mapping= aes(x='alive'))+
              plotnine.labs(title="histogram for titanic passengers")+
              plotnine.geom_histogram()+
              plotnine.geom_rug())
hist
```

```python
kde= (ggplot(data=df,mapping=aes(x='age'))+
            plotnine.geom_density()+
            plotnine.labs(title="Density plot for age of passengers"))
kde
```

```python
bp=(ggplot(data=df,mapping=aes(x="embarked",y="fare"))+
    plotnine.geom_boxplot()+
    plotnine.labs(title="Box plot for Embarked vs Fare"))
bp
```

```python
vr =(ggplot(data=df,mapping =aes(x="sex",y="age"))+
     plotnine.labs(title="Violin plot for passenger age by gender")+
     plotnine.geom_jitter(position=plotnine.position_jitter(0.3)))
vr
```
