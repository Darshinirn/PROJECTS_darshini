# Lets plot the data science salaries  

Exploratory data analysis is a way to better understand data. The numerical data will be visualized in graphs.

Let's begin by importingmatplotlib.pyplot and seaborn. Some graph visualization options are set as well.

## Distribution of Salary (USD) by Experience

we can plot a histogram to know how the salary in usd varies with the expereince so that we can insight about how much experience is needed for each desired salary level

```python
sns.histplot(x = 'salary_in_usd', hue = 'experience_level', multiple = 'stack', bins = 50, data = data_science_salaries_df, palette = 'viridis')
plt.title('Distribution of Salary (USD) by Experience')
plt.xlabel('Salary')
plt.ylabel('Count')
```

From this graph the SE range earns more and they have higher count than other experienced employees

When visualizing the distribution of salary vs experience in the data science salaries 2023 data set, there are several insights that you might find:
Generally, salaries tend to increase as experience increases. This is a common trend in many industries, and is likely due to the fact that more experienced workers are able to bring more value to their employers.
The distribution of salaries tends to be right-skewed, meaning that there are more data points with lower salaries and fewer data points with higher salaries. This could be due to a variety of factors, such as the fact that there are more entry-level data science positions available than senior-level positions.
There may be some outliers in the data set that are skewing the results. It's important to identify and remove any outliers before drawing conclusions about the distribution of salaries vs experience.
There may be differences in salary distribution between different industries or geographic regions. It's important to take these factors into account when analyzing the data.
Overall, visualizing the distribution of salary vs experience in the data science salaries 2023 data set can provide valuable insights into trends in the industry and help employers and employees make informed decisions about compensation.

## Distribution of Salary (USD) by Year

we can plot a histogram to know how the salary in usd varies with the  year so that we can insight about how the effect of change in years affect the salary .

```python
sns.lineplot(y = 'salary_in_usd', x = 'work_year', data =data_science_salaries_df, alpha = 1)
plt.title('Distribution of Salary (USD) by Year')
plt.xlabel('Year')
plt.ylabel('Salary')
```

From the graph it is understood that the salary is low but wide ranged in prior years, but later the salary hike is true but for a reduced range.There may be an overall upward trend in salaries over time. This could be due to factors such as an increasing demand for data scientists, a shortage of qualified candidates, and/or general inflation.There may be fluctuations in salary from year to year. These fluctuations could be due to a variety of factors, such as changes in the economy, changes in the industry, or changes in the data set itself.There may be differences in salary trends between different industries or geographic regions. It's important to take these factors into account when analyzing the data.

## Distribution of Salary (USD) by Remote Ratio

we can plot a histogram to know how the salary in usd varies with the remote ratio.so that we can insight about how much salary can be earned as remote nd otherwise. 

```python
sns.histplot(x = 'salary_in_usd', hue = 'remote_ratio', multiple = 'stack', alpha = 0.9,
             edgecolor = '#cfd0d4', bins = 50, data =data_science_salaries_df, palette = 'viridis')
plt.title('Distribution of Salary (USD) by Remote Ratio')
plt.xlabel('Salary')
plt.ylabel('Count')
```

The employees having remote ratio of 50 are of very less count and the employees having remote ratio of 100 and 0 are of equal count ,but those having 0 ratio have high salary level.There may be a positive correlation between remote ratio and salary. That is, data scientists who work more remotely may tend to have higher salaries than those who work primarily on-site. This could be due to a variety of factors, such as the fact that remote workers may have more specialized skills or be more productive.

## Distribution of Salary (USD) by Company Size


we can plot a histogram to know how the distribution of salary varies with the company size, so that we can insight about how the effect of size of company affect the salary distribution .

```python
sns.histplot(x = 'salary_in_usd', y = 'company_size', data = data_science_salaries_df, palette = 'viridis')
plt.title('Distribution of Salary (USD) by Company Size')
plt.xlabel('Salary')
plt.ylabel('Count')
```

From this heatmap we can understand that the more number of employees are from medium sized companies and only a few are from the large companies that earn more.There may be a positive correlation between company size and salary. That is, data scientists who work at larger companies may tend to have higher salaries than those who work at smaller companies. This could be due to a variety of factors, such as the fact that larger companies may have more resources to invest in their data science teams.There may be differences in salary distribution between different industries or geographic regions, even when controlling for company size. It's important to take these factors into account when analyzing the data.
