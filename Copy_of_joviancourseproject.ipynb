###Lets plot the data science salaries  

Exploratory data analysis is a way to better understand data. The numerical data will be visualized in graphs.

Let's begin by importingmatplotlib.pyplot and seaborn. Some graph visualization options are set as well.

#Salary distribution


```python
data_science_salaries_df.salary_in_usd.plot(title='salary');
data_science_salaries_df.salary_in_usd.plot(kind='line')
plt.xlabel('salary in usd')
```

By using simple line plot we can plot the salary distribution

#Distribution of Salary (USD) by Experience

we can plot a histogram to know how the salary in usd varies with the expereince so that we can insight about how much experience is needed for each desired salary level

```python
sns.histplot(x = 'salary_in_usd', hue = 'experience_level', multiple = 'stack', bins = 50, data = data_science_salaries_df, palette = 'viridis')
plt.title('Distribution of Salary (USD) by Experience')
plt.xlabel('Salary')
plt.ylabel('Count')
```

From this graph the SE range earns more and they have higher count than other experienced employees

#Distribution of Salary (USD) by Year

we can plot a histogram to know how the salary in usd varies with the  year so that we can insight about how the effect of change in years affect the salary .

```python
sns.lineplot(y = 'salary_in_usd', x = 'work_year', data =data_science_salaries_df, alpha = 1)
plt.title('Distribution of Salary (USD) by Year')
plt.xlabel('Year')
plt.ylabel('Salary')
```

From the graph it is understood that the salary is low but wide ranged in prior years, but later the salary hike is true but for a reduced range

#Distribution of Salary (USD) by Remote Ratio

we can plot a histogram to know how the salary in usd varies with the remote ratio.so that we can insight about how much salary can be earned as remote nd otherwise. 

```python
sns.histplot(x = 'salary_in_usd', hue = 'remote_ratio', multiple = 'stack', alpha = 0.9,
             edgecolor = '#cfd0d4', bins = 50, data =data_science_salaries_df, palette = 'viridis')
plt.title('Distribution of Salary (USD) by Remote Ratio')
plt.xlabel('Salary')
plt.ylabel('Count')
```

The employees having remote ratio of 50 are of very less count and the employees having remote ratio of 100 and 0 are of equal count ,but those having 0 ratio have high salary level.

#Distribution of Salary (USD) by Company Size


we can plot a histogram to know how the distribution of salary varies with the company size, so that we can insight about how the effect of size of company affect the salary distribution .

```python
sns.histplot(x = 'salary_in_usd', y = 'company_size', data = data_science_salaries_df, palette = 'viridis')
plt.title('Distribution of Salary (USD) by Company Size')
plt.xlabel('Salary')
plt.ylabel('Count')
```

From this heatmap we can understand that the more number of employees are from medium sized companies and only a few are from the large companies that earn more.
